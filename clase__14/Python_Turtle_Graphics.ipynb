# 1Ô∏è‚É£ **¬øQu√© es Turtle Graphics?**

**Turtle Graphics** es un m√≥dulo de Python que permite crear **dibujos y gr√°ficos** mediante el movimiento de una ‚Äútortuga virtual‚Äù que deja un trazo mientras se desplaza.
### üß© **Imagina**:
- Una tortuga camina sobre una hoja.
- Cada vez que se mueve, **dibuja una l√≠nea**.
- Puedes decirle que avance, gire, cambie de color, o levante el l√°piz.
- Es ideal para iniciar con la **programaci√≥n visual** y la **l√≥gica secuencial**.

# 2Ô∏è‚É£ **Dibujando l√≠neas y movimientos b√°sicos**

Con Turtle podemos controlar el **movimiento**, el **trazo** y la **apariencia** de la tortuga.  
Cada dibujo se crea a partir de tres acciones principales:

1. **Moverse** (`forward()`, `backward()`)  
2. **Girar** (`right()`, `left()`)  
3. **Controlar el l√°piz** (`penup()`, `pendown()`)

### üß≠ **Direcciones y movimientos**

- Por defecto, la tortuga inicia apuntando hacia la **derecha (0¬∞)**.  
- `right(90)` la gira **90¬∞ a la derecha**.  
- `left(90)` la gira **90¬∞ a la izquierda**.  
- Puedes moverla hacia atr√°s con `backward(x)`.

### üñäÔ∏è **Control del l√°piz**

- `penup()` ‚Üí levanta el l√°piz (no dibuja).  
- `pendown()` ‚Üí baja el l√°piz (vuelve a dibujar).  
- `goto(x, y)` ‚Üí mueve la tortuga a coordenadas espec√≠ficas.  

### üé® **Personalizaci√≥n de la tortuga y el trazo**

Puedes personalizar la apariencia de la tortuga y el estilo del dibujo con los siguientes comandos:

| Comando | Descripci√≥n |
|:---------|:-------------|
| `color("color")` | Cambia el color del trazo (por nombre o c√≥digo RGB). |
| `pencolor("color")` | Cambia el color del l√°piz (igual que `color()`). |
| `fillcolor("color")` | Cambia el color de relleno para figuras cerradas. |
| `pensize(x)` | Ajusta el grosor de la l√≠nea. |
| `shape("nombre")` | Cambia la forma de la tortuga (`"turtle"`, `"arrow"`, `"circle"`, etc.). |
| `speed(n)` | Controla la velocidad de movimiento (1‚Äì10 o `"fastest"`). |


````python
#Ejemplo: movimientos y personalizaci√≥n
from ColabTurtlePlus.Turtle import *

initializeTurtle()
color("purple")
pensize(4)
shape("turtle")
speed(6)

# Dibujo en forma de "L"
forward(100)
right(90)
forward(60)

# Mover sin dibujar
penup()
goto(-50, 50)
pendown()

color("orange")
# Dibujar un tri√°ngulo simple
for _ in range(3):
    forward(80)
    left(120)

# Volver al centro y mostrar texto
penup()
goto(0, 0)
pendown()
color("red")
````

# 3Ô∏è‚É£ **Formas geom√©tricas y bucles**

Cuando necesitamos dibujar **figuras con lados iguales** (como tri√°ngulos, cuadrados o hex√°gonos),  
usar un bucle `for` hace el c√≥digo m√°s corto y legible.  

Cada figura se puede pensar como:
- Repetir **un mismo movimiento** (`forward()`)  
- Y luego **un mismo giro** (`right()` o `left()`)  

### üß© **Ejemplo: cuadrado**

Un cuadrado tiene 4 lados de igual longitud y giros de **90¬∞**.

```python
for _ in range(4):
    forward(100)
    right(90)
```
### üß© **Ejemplo: tri√°ngulo equil√°tero**

Un tri√°ngulo equil√°tero tiene 3 lados y giros de 120¬∞.
```python
for _ in range(3):
    forward(100)
    left(120)
```
## üõû **Generalizaci√≥n**

Si una figura tiene **n lados**, cada giro ser√° de **360 / n grados**. Esto nos permite crear un bucle para **cualquier pol√≠gono regular**.

```python
for _ in range(lados):
    forward(longitud)
    right(360 / lados)
```

## üé® **¬øSolo figuras geom√©tricas? ¬°No necesariamente!**

Aunque muchas figuras en Turtle se basan en **pol√≠gonos regulares**, tambi√©n puedes crear formas **no estrictamente geom√©tricas**.  

Basta con **cambiar los √°ngulos o longitudes** dentro del bucle para obtener dise√±os m√°s creativos.

###üí° **Ejemplo: estrella**

En este caso, los giros no son los de un pol√≠gono regular (no suman 360¬∞),

for i in range(5):

## üåÄ **Bucles anidados y patrones decorativos**

Hasta ahora usamos un **solo bucle `for`** para dibujar figuras regulares (cuadrados, tri√°ngulos, etc.).  
Pero podemos usar **bucles anidados** para crear **patrones m√°s elaborados o repetitivos**.

### üìò **Diferencia entre tipos de bucle**

| Tipo | Descripci√≥n | Ejemplo |
|------|--------------|---------|
| **Simple** | Repite un mismo movimiento en una figura. | Dibuja un cuadrado. |
| **Anidado** | Un bucle externo repite una figura completa varias veces, rot√°ndola o cambiando color. | Dibuja varios cuadrados girados entre s√≠. |

###üí° **Ejemplo: bucle simple (no anidado)**

Este c√≥digo usa un solo bucle, pero el √°ngulo de giro hace que el trazo forme un patr√≥n similar a una flor.

```python
for i in range(36):
    forward(100)
    right(100)
```
### üí´ **Ejemplo: bucle anidado**

El primer `for` dibuja un cuadrado, y el segundo repite el cuadrado varias veces, gir√°ndolo un poco cada vez.

```python
for i in range(36):
    for j in range(4):
        forward(100)
        right(90)
    right(10)
```

üí° Esto produce un patr√≥n circular de cuadrados superpuestos ‚Äî un efecto de flor o mandala.

````python
#Demostraci√≥n: bucle anidado (patr√≥n decorativo)
from ColabTurtlePlus.Turtle import *

initializeTurtle()
shape("turtle")
speed(10)
color("purple")

# Bucle anidado: dibuja cuadrados girados
for i in range(36):
    for j in range(4):
        forward(80)
        right(90)
    right(10)

penup()
goto(-80, -160)
pendown()
color("red")
````

# 6Ô∏è‚É£ **Colores y relleno de figuras**

Adem√°s de trazar l√≠neas, Turtle permite **agregar color y relleno** a las figuras.  Esto ayuda a que los dibujos sean m√°s **visuales y creativos**  

## üé® **Comandos principales**

| Comando | Descripci√≥n |
|:---------|:-------------|
| `color("color")` | Cambia el color del trazo (puedes usar nombres o valores RGB). |
| `fillcolor("color")` | Define el color que se usar√° para rellenar una figura cerrada. |
| `begin_fill()` | Indica el inicio del √°rea a rellenar. |
| `end_fill()` | Cierra el √°rea y aplica el relleno con el color seleccionado. |

### üß© **Ejemplo: cuadrado relleno**

1. Se define un color de l√≠nea con `color()`.  
2. Se define el color del relleno con `fillcolor()`.  
3. Se encierra el dibujo entre `begin_fill()` y `end_fill()`.

```python
color("black")
fillcolor("yellow")
begin_fill()
for i in range(4):
    forward(100)
    right(90)
end_fill()
```

````python
#Ejemplo: colores y relleno de figuras
from ColabTurtlePlus.Turtle import *

initializeTurtle()
speed(8)
shape("turtle")

# Cuadrado relleno
color("black")
fillcolor("gold")
begin_fill()
for _ in range(4):
    forward(100)
    right(90)
end_fill()

# Mover a otro lugar y dibujar un tri√°ngulo relleno
penup()
goto(-150, 0)
pendown()

color("darkgreen")
fillcolor("lightgreen")
begin_fill()
for _ in range(3):
    forward(100)
    left(120)
end_fill()

# Dibujar un c√≠rculo
penup()
goto(150, 0)
pendown()

color("purple")
fillcolor("violet")
begin_fill()
circle(50)
end_fill()

penup()
goto(-60, -140)
pendown()
color("red")
````

# 7Ô∏è‚É£ **Uso de funciones Turtle en proyectos creativos**

Hasta ahora, hemos dibujado cada figura paso a paso.  
Pero en proyectos m√°s grandes o dibujos con muchos elementos,  es mejor **usar funciones** para organizar el c√≥digo y reutilizarlo f√°cilmente.

## üß© **Ventajas de usar funciones**

- Permiten **reutilizar** el mismo dibujo varias veces.  
- Hacen el c√≥digo m√°s **limpio y modular**.  
- Permiten combinar figuras peque√±as para crear **escenas completas**.

###üìò **Ejemplo: funci√≥n para un cuadrado**

```python
def cuadrado(lado, color_linea, color_relleno):
    color(color_linea)
    fillcolor(color_relleno)
    begin_fill()
    for _ in range(4):
        forward(lado)
        right(90)
    end_fill()
```
Luego puedes usarla varias veces:

```python
cuadrado(80, "black", "yellow")
penup()
goto(100, 100)
pendown()
cuadrado(50, "red", "orange")
```
### üé® **Composici√≥n creativa**

Combinando funciones simples se pueden construir dibujos m√°s elaborados,
como flores, casas, banderas o patrones decorativos.

````python
#Ejemplo: uso de funciones para componer dibujos
from ColabTurtlePlus.Turtle import *

initializeTurtle()
speed(8)
shape("turtle")

# Definici√≥n de funciones
def cuadrado(lado, color_linea, color_relleno):
    color(color_linea)
    fillcolor(color_relleno)
    begin_fill()
    for _ in range(4):
        forward(lado)
        right(90)
    end_fill()

def tri√°ngulo(lado, color_linea, color_relleno):
    color(color_linea)
    fillcolor(color_relleno)
    begin_fill()
    for _ in range(3):
        forward(lado)
        left(120)
    end_fill()

def casa():
    cuadrado(100, "brown", "peru")
    tri√°ngulo(100, "black", "darkred")

# Dibujo compuesto
penup()
goto(-50, -50)
pendown()
casa()

penup()
goto(150, -50)
pendown()
casa()

penup()
goto(-60, -160)
pendown()
color("blue")
````

````python
#@title **Instrucciones:**
#@markdown Observa el siguiente c√≥digo y **predice qu√© dibujo formar√°** antes de ejecutarlo.

#@markdown Piensa: ¬øQu√© figura dibuja cada funci√≥n? y ¬øQu√© efecto se crea al combinar colores y repeticiones?
#@markdown
#@markdown ```python
#@markdown from ColabTurtlePlus.Turtle import *
#@markdown initializeTurtle()
#@markdown speed(8)
#@markdown
#@markdown def trazo_vertical():
#@markdown     color("forestgreen")
#@markdown     pensize(4)
#@markdown     right(90)
#@markdown     forward(60)
#@markdown     left(90)
#@markdown
#@markdown def figura_central(tono):
#@markdown     color(tono)
#@markdown     for i in range(6):
#@markdown         circle(20, 60)
#@markdown         left(120)
#@markdown
#@markdown def franja_inferior():
#@markdown     color("limegreen")
#@markdown     pensize(6)
#@markdown     penup()
#@markdown     goto(-200, -100)
#@markdown     pendown()
#@markdown     forward(400)
#@markdown
#@markdown franja_inferior()
#@markdown
#@markdown penup(); goto(-100, -40); pendown()
#@markdown trazo_vertical(); figura_central("deeppink")
#@markdown
#@markdown penup(); goto(0, -40); pendown()
#@markdown trazo_vertical(); figura_central("gold")
#@markdown
#@markdown penup(); goto(100, -40); pendown()
#@markdown trazo_vertical(); figura_central("orange")
#@markdown ```
Prediccion = ""  #@param {type:"string"}

from ColabTurtlePlus.Turtle import *

initializeTurtle()
speed(8)

def franja_inferior():
    color("limegreen"); pensize(6)
    penup(); goto(-220, -100); pendown()
    forward(440)

def trazo_vertical(altura=60):
    color("forestgreen"); pensize(4)
    setheading(90)
    pendown()
    forward(altura)

def figura_central(tono, figures=6, radio=20, ang=60):
    pencolor(tono); color(tono); pensize(2)
    x, y = position()
    for k in range(figures):
        setheading(90 + k * (360 / figures))
        penup(); goto(x, y); pendown()
        circle(radio, ang)
        left(180 - ang)
        circle(radio, ang)

# Escena
franja_inferior()

# Flor 1
penup(); goto(-120, -100); pendown()
trazo_vertical(70); figura_central("deeppink")

# Flor 2
penup(); goto(0, -100); pendown()
trazo_vertical(70); figura_central("gold")

# Flor 3
penup(); goto(120, -100); pendown()
trazo_vertical(70); figura_central("orange")

print("\nTu predicci√≥n:")
print(Prediccion if Prediccion.strip() else "(sin respuesta)")
````

# üí° **Tips y recomendaciones**

### üß≠ **Uso general**
- Antes de cada nuevo dibujo, usa `initializeTurtle()` para **reiniciar el lienzo**.  
- Ajusta la velocidad con `speed(n)` (1‚Äì10) para observar el movimiento o hacerlo m√°s r√°pido.  
- Si la tortuga se desorienta, usa `setheading(0)` para restablecerla apuntando a la derecha.  
- Combina `penup()`, `pendown()` y `goto(x, y)` para moverte sin dejar trazos.
- Si el dibujo se sale del lienzo, prueba con `reset()` o redimensiona el √°rea de dibujo (`initializeTurtle(canvas_width=600, canvas_height=400)`).
