# 1Ô∏è‚É£ **Introducci√≥n a Matplotlib**

**Matplotlib** es una de las bibliotecas m√°s importantes de Python para la **visualizaci√≥n de datos**, pues permite transformar informaci√≥n num√©rica en **gr√°ficos claros, atractivos y f√°ciles de interpretar**.  

Aunque Python puede trabajar con datos directamente desde listas o arreglos NumPy, muchas veces los n√∫meros por s√≠ solos no permiten ver patrones, tendencias o relaciones de forma intuitiva.  

Aqu√≠ es donde Matplotlib resulta fundamental: ofrece herramientas para crear **l√≠neas, gr√°ficos de barras, dispersiones, histogramas y muchas otras formas de representaci√≥n visual**, ayudando a entender y comunicar mejor los datos.

## üìò **Concepto b√°sico**

Matplotlib se basa en un m√≥dulo llamado `pyplot`, que funciona como una ‚Äúcaja de herramientas‚Äù para crear gr√°ficos paso a paso. La idea principal es muy simple:

1. Preparas tus **datos** (por ejemplo, listas o arreglos NumPy).  
2. Llamas a **funciones de dibujo** como `plt.plot()`, `plt.scatter()`, `plt.bar()`, etc., para visualizarlos.  

Cada gr√°fico se construye a√±adiendo elementos como **l√≠neas, puntos, etiquetas, t√≠tulos y colores**, lo que hace que entender la informaci√≥n sea mucho m√°s f√°cil que leer n√∫meros directamente.


## üß© **Comandos esenciales**
| Comando | Descripci√≥n |
|:---------|:-------------|
| `plt.plot(x, y)` | Crea un **gr√°fico de l√≠nea** que une los puntos definidos por `x` y `y`. |
| `plt.scatter(x, y)` | Dibuja puntos individuales, formando un **gr√°fico de dispersi√≥n**. |
| `plt.bar(x, y)` | Crea un **gr√°fico de barras**, √∫til para comparar valores entre categor√≠as. |
| `plt.xlabel("texto")` / `plt.ylabel("texto")` | A√±aden etiquetas descriptivas a los ejes X y Y para que el gr√°fico sea entendible. |
| `plt.title("texto")` | Agrega un t√≠tulo al gr√°fico. |
| `plt.legend()` | Muestra una **leyenda** cuando se tienen varias l√≠neas o elementos con etiquetas. |
| `plt.grid(True)` | Activa una cuadr√≠cula para facilitar la lectura visual de valores. |
| `plt.figure(figsize=(a,b))` | Crea una figura con un tama√±o personalizado (ancho y alto en pulgadas). |
| `plt.show()` | Renderiza y muestra el gr√°fico en pantalla. |

üí° *Idea clave:*  
Matplotlib se comporta como un ‚Äúdibujante‚Äù: cada funci√≥n agrega algo a la imagen final. Por eso puedes construir tu gr√°fico **paso a paso**, a√±adiendo l√≠neas, etiquetas, colores y estilos seg√∫n lo necesites.

````python
#**Ejemplo gr√°fico lineal** üìà
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
x = np.arange(0, 11)
y = x ** 2

# Creaci√≥n del gr√°fico
plt.figure(figsize=(7,4))
plt.plot(x, y, color="royalblue", linewidth=2, marker="o", label="y = x¬≤")

# Personalizaci√≥n
plt.title("Gr√°fico lineal b√°sico")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend()
plt.grid(True)

plt.show()
````

# 2Ô∏è‚É£ **Tipos b√°sicos de gr√°ficos**

Matplotlib permite crear distintos tipos de gr√°ficos seg√∫n el tipo de informaci√≥n que desees visualizar. Cada uno resalta un aspecto diferente de los datos, como tendencias, comparaciones o distribuciones.

## üìà **1. Gr√°fico de l√≠neas (`plt.plot`)**

Ideal para visualizar **tendencias en el tiempo** o la evoluci√≥n de una variable.

Se interpreta f√°cilmente la subida, bajada o estabilidad de los valores.

````python
#Ejemplo: Gr√°fico de l√≠nea üìâ
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 10)
y = np.sin(x)

plt.figure(figsize=(7,4))
plt.plot(x, y, marker="o", color="blue", label="sin(x)")
plt.title("Gr√°fico de l√≠nea")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend()
plt.grid(True)
plt.show()
````

## üìä **2. Gr√°fico de barras (`plt.bar`)**

Sirve para comparar **categor√≠as** o grupos.

Ejemplos comunes:
- Ventas por producto
- Poblaci√≥n por pa√≠s
- Votos por candidato

````python
#Ejemplo: Gr√°fico de barras üìä
import matplotlib.pyplot as plt

categorias = ["A", "B", "C", "D"]
valores = [30, 12, 45, 10]

plt.figure(figsize=(7,4))
plt.bar(categorias, valores, color="green")
plt.title("Gr√°fico de barras")
plt.xlabel("Categor√≠as")
plt.ylabel("Valores")
plt.grid(axis="y")
plt.show()
````

## üîµ **3. Gr√°fico de dispersi√≥n (`plt.scatter`)**

Muestra **pares de valores como puntos**, perfecto para analizar **relaciones** entre dos variables.

Ejemplos:
- Tama√±o vs peso
- Horas de estudio vs calificaci√≥n

````python
#Ejemplo: Gr√°fico de dispersi√≥n üîµ
import numpy as np
import matplotlib.pyplot as plt

# Generamos una relaci√≥n sencilla: y aumenta ligeramente con x
x = np.linspace(0, 10, 20)
y = 0.5 * x + np.random.randn(20) * 1.2   # relaci√≥n con un poco de ruido

plt.figure(figsize=(7,4))
plt.scatter(x, y, color="teal", s=60, edgecolors="black")
plt.title("Gr√°fico de dispersi√≥n")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.grid(True)
plt.show()
````

## ü•ß **4. Gr√°fico circular o de pastel (`plt.pie`)**

Representa **porcentajes o proporciones** de un total.

Es √∫til cuando quieres mostrar c√≥mo se divide un conjunto en partes.

````python
#Ejemplo: Gr√°fico circular ü•ß
import matplotlib.pyplot as plt

etiquetas = ["Python", "Java", "C++", "JavaScript"]
porcentajes = [40, 25, 20, 15]

plt.figure(figsize=(6,6))
plt.pie(porcentajes, labels=etiquetas, autopct="%1.1f%%", startangle=90)
plt.title("Distribuci√≥n de lenguajes")
plt.show()
````

# 3Ô∏è‚É£ **Personalizaci√≥n de gr√°ficos**

Una de las ventajas de Matplotlib es la cantidad de formas en las que puedes personalizar un gr√°fico.  
Esto permite que la **visualizaci√≥n sea m√°s clara, m√°s est√©tica y m√°s adecuada al tipo de datos** que est√°s mostrando.

Aqu√≠ aprenderemos a modificar:

- Colores üé®  
- Estilos de l√≠nea  
- Marcadores (puntos)  
- Grosor de l√≠neas  
- Tama√±o de figuras  
- T√≠tulos y etiquetas  
- Leyendas  
- Cuadr√≠culas  

Cada uno de estos elementos mejora la **interpretaci√≥n visual** del gr√°fico.

## üé® **1. Colores y estilos de l√≠nea**

Puedes usar colores por nombre (`"red"`), c√≥digos hex (`"#FF5733"`), RGB, o listas predefinidas.

**Estilos de l√≠nea** comunes:
- `"-"` ‚Üí l√≠nea continua  
- `"--"` ‚Üí l√≠nea discontinua  
- `"-."` ‚Üí l√≠nea punto-guion  
- `":"` ‚Üí l√≠nea punteada  

## üîµ **2. Marcadores**

Los marcadores resaltan los puntos en un gr√°fico de l√≠nea o dispersi√≥n.

Ejemplos:
- `"o"` ‚Üí c√≠rculo  
- `"s"` ‚Üí cuadrado  
- `"^"` ‚Üí tri√°ngulo  
- `"x"` ‚Üí equis  

## ‚úèÔ∏è **3. Grosor y tama√±o**

- `linewidth=x` ‚Üí grosor de l√≠nea  
- `markersize=x` ‚Üí tama√±o de los marcadores  
- `figsize=(ancho, alto)` ‚Üí tama√±o de la figura  

## üè∑Ô∏è **4. T√≠tulos y etiquetas**

- `plt.title("Mi gr√°fico")`  
- `plt.xlabel("Tiempo (s)")`  
- `plt.ylabel("Amplitud")`  

Un gr√°fico sin etiquetas es dif√≠cil de interpretar.

## üß≠ **5. Leyenda**

Cuando tienes varias l√≠neas o series de datos:

```python
plt.plot(x, y1, label="Serie 1")
plt.plot(x, y2, label="Serie 2")
plt.legend()
```
## üìà **6. Cuadr√≠cula**

La cuadr√≠cula ayuda a leer los valores:

```python
plt.grid(True)
```

````python
#Ejemplo: Personalizaci√≥n de gr√°ficos üñåÔ∏è

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 50)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(8,5))

# Primera l√≠nea
plt.plot(
    x, y1,
    color="crimson",
    linestyle="--",
    linewidth=2.5,
    marker="o",
    markersize=6,
    label="sin(x)"
)

# Segunda l√≠nea
plt.plot(
    x, y2,
    color="royalblue",
    linestyle="-.",
    linewidth=2,
    marker="s",
    markersize=5,
    label="cos(x)"
)

# Personalizaci√≥n visual
plt.title("Ejemplo de Personalizaci√≥n de Gr√°ficos")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid(True)
plt.legend()

plt.show()
````

# 4Ô∏è‚É£ **Procesamiento matem√°tico de datos**

Para analizar datos necesitamos aplicar **operaciones matem√°ticas b√°sicas**: promedios, sumas, diferencias, m√°ximos, m√≠nimos, transformaciones, normalizaci√≥n y m√°s.

Python ofrece herramientas muy poderosas para esto, especialmente a trav√©s de **NumPy**, con la cual ya est√°s familiarizado, permitiendo trabajar con datos de forma r√°pida y eficiente.

Estas operaciones son fundamentales antes de graficar, porque ayudan a:
- Resumir informaci√≥n  
- Limpiar datos ruidosos  
- Detectar tendencias  
- Preparar datos para ser visualizados correctamente

## üßÆ **Operaciones num√©ricas esenciales con NumPy**

| Operaci√≥n | Ejemplo | Descripci√≥n |
|----------|---------|-------------|
| `np.mean(x)` | promedio | Mide tendencia central |
| `np.sum(x)` | suma total | Suma todos los elementos |
| `np.max(x)` / `np.min(x)` | m√°ximos/m√≠nimos | Encuentra valores extremos |
| `np.std(x)` | desviaci√≥n est√°ndar | Mide variabilidad |
| `np.diff(x)` | diferencias entre puntos | √ötil para ver cambios |
| `(x - min)/(max - min)` | normalizaci√≥n | Escala datos 0‚Äì1 |

````python
#Ejemplo: Normalizaci√≥n de datos y su gr√°fica
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 50)
y = np.sin(x) + np.random.randn(50)*0.3  # se√±al con ruido

# Normalizaci√≥n 0‚Äì1
y_norm = (y - y.min()) / (y.max() - y.min())

plt.figure(figsize=(7,4))
plt.plot(x, y_norm, color="darkcyan", marker="o")
plt.title("Datos normalizados (0‚Äì1)")
plt.grid(True)
plt.show()
````

# 5Ô∏è‚É£ **An√°lisis y visualizaci√≥n combinada**

Es muy com√∫n combinar **operaciones matem√°ticas** con **gr√°ficos**, porque visualizar los resultados te permite identificar patrones, comparar comportamientos y entender mejor lo que est√° pasando.

A continuaci√≥n ver√°s tres an√°lisis muy frecuentes cuando se trabaja con datos num√©ricos.

## üß† **¬øC√≥mo se hacen estos an√°lisis?**

Cada uno sigue la misma l√≥gica general:

1. Tienes datos originales.  
2. Les aplicas una **transformaci√≥n matem√°tica** (suavizado, tendencia, cambio de funci√≥n‚Ä¶).  
3. Visualizas las versiones para compararlas.

### üíª **Suavizado (reducci√≥n de ruido)**

Sirve para **eliminar variaciones bruscas** y observar la **forma general** de los datos.

**C√≥mo se hace:**

```python
k = 5   # tama√±o de la ventana
y_suave = np.convolve(y, np.ones(k)/k, mode="same")
```
Esto aplica un promedio m√≥vil y genera una versi√≥n m√°s ‚Äúlimpia‚Äù.

### üíª **Tendencia lineal (comportamiento general)**

Muestra si los **datos suben, bajan, o se mantienen estables** en promedio.

C√≥mo se hace:

```python
m, b = np.polyfit(x, y, 1)
tendencia = m*x + b
```
Ese c√°lculo **ajusta una l√≠nea recta que representa la direcci√≥n general** del conjunto.

### üíª **Transformaciones matem√°ticas**

Permiten comparar **c√≥mo cambia una se√±al cuando le aplicas distintas funciones**.

C√≥mo se hace:

```python
y1 = np.sin(x)
y2 = np.sin(2*x)
y3 = np.sin(x) + 0.5*np.cos(3*x)
```
Cada curva tendr√° frecuencias, amplitudes o formas diferentes, y verlas juntas revela sus diferencias.

##üíª **Secci√≥n de Ejemplos**

````python
#@title **An√°lisis: Datos originales vs Suavizados**
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 60)
y = np.sin(x) + np.random.randn(60)*0.4

k = 5
y_suave = np.convolve(y, np.ones(k)/k, mode="same")

plt.figure(figsize=(8,4))
plt.plot(x, y, alpha=0.4, label="Original (azul)", color="royalblue")
plt.plot(x, y_suave, linewidth=3, color="crimson", label="Suavizado (roja)")
plt.title("Datos originales vs suavizados")
plt.legend()
plt.grid(True)
plt.show()
````

````python
#@title **An√°lisis: L√≠nea de tendencia**
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 20)
y = 3*x + np.random.randn(20)*5

m, b = np.polyfit(x, y, 1)
tendencia = m*x + b

plt.figure(figsize=(7,4))
plt.scatter(x, y, color="royalblue", label="Datos")
plt.plot(x, tendencia, color="darkorange", linewidth=3, label="Tendencia")
plt.title("Datos vs tendencia lineal")
plt.legend()
plt.grid(True)
plt.show()
````

````python
#@title **An√°lisis: Comparaci√≥n de transformaciones**
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4*np.pi, 200)
y1 = np.sin(x)
y2 = np.sin(2*x)
y3 = np.sin(x) + 0.5*np.cos(3*x)

plt.figure(figsize=(8,4))
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="sin(2x)")
plt.plot(x, y3, label="sin(x) + 0.5cos(3x)")
plt.title("Comparaci√≥n de transformaciones")
plt.legend()
plt.grid(True)
plt.show()
````

````python
#@title **Instrucciones**:
#@markdown Ejecuta el c√≥digo, observa la gr√°fica y responde **¬øQu√© transformaci√≥n corresponde a cada l√≠nea?**

Respuesta = ""  #@param {type:"string"}

import numpy as np
import matplotlib.pyplot as plt

# Datos base con ruido
x = np.linspace(0, 10, 120)
y = np.sin(x) + np.random.randn(120)*0.35

# Transformaci√≥n A: suavizado
k = 8
y_A = np.convolve(y, np.ones(k)/k, mode="same")

# Transformaci√≥n B: versi√≥n trigonom√©trica modificada
y_B = np.sin(2*x) * 0.5 + np.random.randn(120)*0.1

plt.figure(figsize=(8,4))
plt.plot(x, y, label="Original (azul)", alpha=0.3)
plt.plot(x, y_A, label="A (roja)", color="crimson", linewidth=2)
plt.plot(x, y_B, label="B (verde)", color="seagreen", linewidth=2)
plt.legend()
plt.grid(True)
plt.show()

print("\nTu respuesta:")
print(Respuesta if Respuesta.strip() else "(sin respuesta)")
````

# üí° **Tips y recomendaciones**

## üß≠ **Tips para trabajar con datos**

- Usa NumPy siempre que puedas: es **m√°s r√°pido** y **m√°s preciso** que las listas de Python.
- Antes de graficar, revisa si tus datos necesitan:
  - Limpieza (valores at√≠picos)
  - Normalizaci√≥n (0‚Äì1)
  - Suavizado (ruido)
  - Transformaci√≥n (logar√≠tmica, trigonom√©trica, etc.)
- Combina an√°lisis num√©rico con gr√°ficos: no basta el n√∫mero, ni basta el gr√°fico por s√≠ solo.
- Prueba varias escalas: lineal vs. logar√≠tmica puede cambiar completamente la interpretaci√≥n.
- Usa `np.polyfit()` para tendencias y compararlas con los datos reales.

## üé® **Tips para visualizaci√≥n con Matplotlib**

- Ajusta `figsize=(ancho, alto)` para que tus gr√°ficos sean m√°s claros.
- Usa colores contrastantes cuando compares varias curvas.
- Evita saturar el gr√°fico: mejor **menos elementos, mejor organizados**.
- La leyenda siempre debe ser clara y corta.
- Agrega t√≠tulos y etiquetas incluso si parece obvio: facilita la lectura.
- Usa `alpha` (transparencia) para superponer datos sin confundir.
- Si algo no se ve bien, prueba cambiar:  
  - Tipo de l√≠nea (`--`, `:`, `-.`)
  - Grosor (`linewidth`)
  - Marcadores (`marker="o"`, `"s"`, `"^"`)

## ‚öôÔ∏è **Recomendaciones pr√°cticas**

- Si tu figura queda muy peque√±a, usa `plt.figure(figsize=(8,5))`.
- Para varios gr√°ficos juntos usa `plt.subplot()` o `plt.subplots()`.
- Si el gr√°fico se ‚Äúencima‚Äù de las etiquetas, llama `plt.tight_layout()`.
